This involves defining functions for simulating sensor readings, checking ranges, and a main loop for fault detection, followed by a loop to run the simulation. All these steps can be combined into a single code block.



```python
import random
import time

def simulate_sensor_reading(normal_range=(100, 110), fault_probability=0.1, fault_value=150):
  """Simulates reading a sensor value, occasionally returning a fault value."""
  if random.random() < fault_probability:
    return fault_value  # Simulate a fault
  else:
    return random.uniform(normal_range[0], normal_range[1]) # Normal reading

def check_sensor_range(reading, acceptable_range=(95, 115)):
  """Checks if a sensor reading is within an acceptable range."""
  return acceptable_range[0] <= reading <= acceptable_range[1]

def fault_detection_loop(iterations=5, normal_range=(100, 110), acceptable_range=(95, 115)):
  """Simulates the main loop of a fault detection system."""
  print("\nStarting fault detection simulation...")
  for i in range(iterations):
    sensor_value = simulate_sensor_reading(normal_range)
    is_within_range = check_sensor_range(sensor_value, acceptable_range)

    print(f"Iteration {i+1}: Sensor reading = {sensor_value:.2f}")

    if not is_within_range:
      print("ALERT: Sensor reading out of acceptable range!")
    time.sleep(0.5) # Simulate a delay between readings
  print("Fault detection simulation finished.")

# Run the simulation
fault_detection_loop(iterations=10)
```

```python
import random
import time
import math

def simulate_imu_reading(normal_accel=(0, 0, -9.81), accel_noise=0.1, accel_stuck_fault_prob=0.05, accel_noisy_fault_prob=0.03, accel_failure_prob=0.01,
                         normal_gyro=(0, 0, 0), gyro_noise=0.01, gyro_stuck_fault_prob=0.05, gyro_noisy_fault_prob=0.03, gyro_failure_prob=0.01):
  """Simulates IMU (accelerometer and gyroscope) readings with different fault types."""
  fault_status = {"accelerometer": "normal", "gyroscope": "normal"}
  accel_reading = list(normal_accel)
  gyro_reading = list(normal_gyro)

  # Simulate accelerometer faults
  if random.random() < accel_failure_prob:
    accel_reading = None
    fault_status["accelerometer"] = "failure"
  elif random.random() < accel_stuck_fault_prob:
    stuck_axis = random.randint(0, 2)
    accel_reading[stuck_axis] = normal_accel[stuck_axis] # Simulate stuck at normal value for simplicity
    fault_status["accelerometer"] = f"stuck_axis_{stuck_axis}"
  elif random.random() < accel_noisy_fault_prob:
    accel_reading = [a + random.uniform(-accel_noise * 5, accel_noise * 5) for a in accel_reading] # Increased noise
    fault_status["accelerometer"] = "noisy"
  else:
    accel_reading = [a + random.uniform(-accel_noise, accel_noise) for a in accel_reading] # Normal noise

  # Simulate gyroscope faults
  if random.random() < gyro_failure_prob:
    gyro_reading = None
    fault_status["gyroscope"] = "failure"
  elif random.random() < gyro_stuck_fault_prob:
    stuck_axis = random.randint(0, 2)
    gyro_reading[stuck_axis] = normal_gyro[stuck_axis] # Simulate stuck at normal value for simplicity
    fault_status["gyroscope"] = f"stuck_axis_{stuck_axis}"
  elif random.random() < gyro_noisy_fault_prob:
    gyro_reading = [g + random.uniform(-gyro_noise * 10, gyro_noise * 10) for g in gyro_reading] # Increased noise
    fault_status["gyroscope"] = "noisy"
  else:
    gyro_reading = [g + random.uniform(-gyro_noise, gyro_noise) for g in gyro_reading] # Normal noise


  return accel_reading, gyro_reading, fault_status

def simulate_gps_reading(normal_lat=37.7749, normal_lon=-122.4194, position_noise=0.0001, drift_rate=0.00001,
                         sudden_jump_prob=0.02, loss_of_signal_prob=0.01, current_lat=None, current_lon=None):
  """Simulates GPS (latitude and longitude) readings with different fault types."""
  fault_status = "normal"

  if current_lat is None or current_lon is None:
      current_lat = normal_lat
      current_lon = normal_lon

  # Simulate faults
  if random.random() < loss_of_signal_prob:
    lat_reading = None
    lon_reading = None
    fault_status = "loss_of_signal"
  elif random.random() < sudden_jump_prob:
    lat_reading = current_lat + random.uniform(-0.1, 0.1) # Sudden jump
    lon_reading = current_lon + random.uniform(-0.1, 0.1) # Sudden jump
    fault_status = "sudden_jump"
  else:
    # Simulate normal reading with noise and drift
    lat_reading = current_lat + random.uniform(-position_noise, position_noise) + drift_rate
    lon_reading = current_lon + random.uniform(-position_noise, position_noise) + drift_rate

  return lat_reading, lon_reading, fault_status

def simulate_airspeed_reading(normal_airspeed=15, airspeed_noise=0.5, stuck_fault_prob=0.05, intermittent_prob=0.03, failure_prob=0.01):
  """Simulates airspeed readings with different fault types."""
  fault_status = "normal"
  airspeed_reading = normal_airspeed

  # Simulate faults
  if random.random() < failure_prob:
    airspeed_reading = None
    fault_status = "failure"
  elif random.random() < stuck_fault_prob:
    airspeed_reading = normal_airspeed # Simulate stuck at normal value for simplicity
    fault_status = "stuck"
  elif random.random() < intermittent_prob:
    if random.random() < 0.5: # Intermittently return None
      airspeed_reading = None
      fault_status = "intermittent"
    else:
        airspeed_reading = normal_airspeed + random.uniform(-airspeed_noise, airspeed_noise) # Normal reading on some iterations
  else:
    airspeed_reading = normal_airspeed + random.uniform(-airspeed_noise, airspeed_noise) # Normal noise

  return airspeed_reading, fault_status

# Example usage (for testing)
# accel, gyro, imu_fault = simulate_imu_reading()
# print(f"IMU Reading: Accel={accel}, Gyro={gyro}, Fault Status={imu_fault}")

# lat, lon, gps_fault = simulate_gps_reading()
# print(f"GPS Reading: Lat={lat}, Lon={lon}, Fault Status={gps_fault}")

# airspeed, airspeed_fault = simulate_airspeed_reading()
# print(f"Airspeed Reading: {airspeed}, Fault Status={airspeed_fault}")
```

```python
from collections import deque
import statistics

# Function to check rate of change
def check_rate_of_change(current_reading, historical_readings, time_interval=0.1, threshold=5.0):
  """Checks if the rate of change of a sensor reading exceeds a threshold."""
  if len(historical_readings) < 1 or current_reading is None or historical_readings[-1] is None:
    return False # Not enough data or current/previous reading is None

  previous_reading = historical_readings[-1]
  rate_of_change = abs((current_reading - previous_reading) / time_interval)

  return rate_of_change > threshold

# Function to check consistency between related sensors (placeholder)
def check_consistency(sensor1_reading, sensor2_reading, threshold=10.0):
  """Checks for significant discrepancy between two related sensor readings."""
  if sensor1_reading is None or sensor2_reading is None:
    return False # Cannot check consistency if either reading is None

  return abs(sensor1_reading - sensor2_reading) > threshold

# Function to check for statistical anomaly (using moving average)
def check_statistical_anomaly(current_reading, historical_readings, window_size=5, threshold_multiplier=2.0):
  """Checks if a sensor reading is statistically anomalous based on a moving average."""
  if len(historical_readings) < window_size or current_reading is None:
    return False # Not enough data or current reading is None

  recent_readings = [r for r in list(historical_readings)[-window_size:] if r is not None]

  if len(recent_readings) < window_size:
      return False # Not enough valid data points in the window

  moving_avg = statistics.mean(recent_readings)
  moving_stddev = statistics.stdev(recent_readings) if len(recent_readings) > 1 else 0

  # Avoid division by zero if standard deviation is zero
  if moving_stddev == 0:
      return abs(current_reading - moving_avg) > 0 # Flag if different from constant value
  else:
      z_score = abs((current_reading - moving_avg) / moving_stddev)
      return z_score > threshold_multiplier

# Example usage (for testing) - assuming historical_readings is a deque
# imu_history = deque([10, 10.1, 10.2], maxlen=10)
# current_imu = 15
# print(f"Rate of change anomaly: {check_rate_of_change(current_imu, list(imu_history))}")
# print(f"Statistical anomaly: {check_statistical_anomaly(current_imu, imu_history)}")

# gps_alt_history = deque([100, 101, 102], maxlen=10)
# baro_alt_history = deque([102, 103, 104], maxlen=10)
# current_gps_alt = 105
# current_baro_alt = 115
# print(f"Consistency anomaly: {check_consistency(current_gps_alt, current_baro_alt)}")
```

```python
import pandas as pd
import numpy as np
import random
import time

def generate_normal_data(duration=100, sampling_rate=10, normal_accel=(0, 0, -9.81), accel_noise=0.1,
                         normal_gyro=(0, 0, 0), gyro_noise=0.01,
                         normal_lat=37.7749, normal_lon=-122.4194, position_noise=0.0001, drift_rate=0.00001,
                         normal_airspeed=15, airspeed_noise=0.5):
    """Generates normal sensor readings over time."""
    num_steps = int(duration * sampling_rate)
    time_index = pd.to_timedelta(np.arange(num_steps) / sampling_rate, unit='s')

    data = {
        'time': time_index,
        'accel_x': np.random.normal(normal_accel[0], accel_noise, num_steps),
        'accel_y': np.random.normal(normal_accel[1], accel_noise, num_steps),
        'accel_z': np.random.normal(normal_accel[2], accel_noise, num_steps),
        'gyro_x': np.random.normal(normal_gyro[0], gyro_noise, num_steps),
        'gyro_y': np.random.normal(normal_gyro[1], gyro_noise, num_steps),
        'gyro_z': np.random.normal(normal_gyro[2], gyro_noise, num_steps),
        'gps_lat': normal_lat + np.cumsum(np.random.normal(-position_noise, position_noise, num_steps) + drift_rate),
        'gps_lon': normal_lon + np.cumsum(np.random.normal(-position_noise, position_noise, num_steps) + drift_rate),
        'airspeed': np.random.normal(normal_airspeed, airspeed_noise, num_steps)
    }

    df = pd.DataFrame(data).set_index('time')
    return df

def inject_fault(data, sensor_type, fault_type, start_time, end_time, fault_value=None, noise_multiplier=None, axis=None):
    """Injects a specific fault into the data within a defined time window."""
    df_faulted = data.copy()
    fault_mask = (df_faulted.index >= pd.to_timedelta(start_time, unit='s')) & \
                 (df_faulted.index <= pd.to_timedelta(end_time, unit='s'))

    if sensor_type == 'imu_accel' and axis is not None:
        col_name = f'accel_{axis}'
        if fault_type == 'stuck':
            df_faulted.loc[fault_mask, col_name] = df_faulted.loc[fault_mask, col_name].mean() if fault_value is None else fault_value
        elif fault_type == 'noisy' and noise_multiplier is not None:
            df_faulted.loc[fault_mask, col_name] = df_faulted.loc[fault_mask, col_name] * noise_multiplier
        elif fault_type == 'failure':
            df_faulted.loc[fault_mask, col_name] = np.nan # Represent complete failure with NaN

    elif sensor_type == 'imu_gyro' and axis is not None:
        col_name = f'gyro_{axis}'
        if fault_type == 'stuck':
             df_faulted.loc[fault_mask, col_name] = df_faulted.loc[fault_mask, col_name].mean() if fault_value is None else fault_value
        elif fault_type == 'noisy' and noise_multiplier is not None:
            df_faulted.loc[fault_mask, col_name] = df_faulted.loc[fault_mask, col_name] * noise_multiplier
        elif fault_type == 'failure':
            df_faulted.loc[fault_mask, col_name] = np.nan

    elif sensor_type == 'gps_lat':
        if fault_type == 'stuck':
             df_faulted.loc[fault_mask, 'gps_lat'] = df_faulted.loc[fault_mask, 'gps_lat'].mean() if fault_value is None else fault_value
        elif fault_type == 'sudden_jump':
            df_faulted.loc[fault_mask, 'gps_lat'] = df_faulted.loc[fault_mask, 'gps_lat'] + (fault_value if fault_value is not None else 0.1) # Add a jump
        elif fault_type == 'loss_of_signal':
             df_faulted.loc[fault_mask, 'gps_lat'] = np.nan

    elif sensor_type == 'gps_lon':
        if fault_type == 'stuck':
             df_faulted.loc[fault_mask, 'gps_lon'] = df_faulted.loc[fault_mask, 'gps_lon'].mean() if fault_value is None else fault_value
        elif fault_type == 'sudden_jump':
            df_faulted.loc[fault_mask, 'gps_lon'] = df_faulted.loc[fault_mask, 'gps_lon'] + (fault_value if fault_value is not None else 0.1) # Add a jump
        elif fault_type == 'loss_of_signal':
             df_faulted.loc[fault_mask, 'gps_lon'] = np.nan

    elif sensor_type == 'airspeed':
        if fault_type == 'stuck':
            df_faulted.loc[fault_mask, 'airspeed'] = df_faulted.loc[fault_mask, 'airspeed'].mean() if fault_value is None else fault_value
        elif fault_type == 'intermittent':
            # Intermittently set to NaN within the fault window
            intermittent_mask = df_faulted.index[fault_mask]
            nan_indices = random.sample(list(intermittent_mask), int(len(intermittent_mask) * 0.5)) # 50% intermittent failure
            df_faulted.loc[nan_indices, 'airspeed'] = np.nan
        elif fault_type == 'failure':
             df_faulted.loc[fault_mask, 'airspeed'] = np.nan


    return df_faulted

def generate_dataset(num_normal_periods=5, normal_duration=50, fault_duration=10, sampling_rate=10):
    """Generates a dataset with periods of normal operation and injected faults."""
    all_data = []
    all_labels = []
    fault_types = {
        'imu_accel': ['stuck', 'noisy', 'failure'],
        'imu_gyro': ['stuck', 'noisy', 'failure'],
        'gps_lat': ['stuck', 'sudden_jump', 'loss_of_signal'],
        'gps_lon': ['stuck', 'sudden_jump', 'loss_of_signal'],
        'airspeed': ['stuck', 'intermittent', 'failure']
    }
    axes = ['x', 'y', 'z']

    current_time = 0

    for i in range(num_normal_periods):
        # Generate normal data
        normal_df = generate_normal_data(duration=normal_duration, sampling_rate=sampling_rate)
        normal_df.index = normal_df.index + pd.to_timedelta(current_time, unit='s')
        all_data.append(normal_df)
        all_labels.append(pd.Series('normal', index=normal_df.index))
        current_time += normal_duration

        # Inject a random fault after each normal period (except possibly the last)
        if i < num_normal_periods - 1:
            sensor_type = random.choice(list(fault_types.keys()))
            fault_type = random.choice(fault_types[sensor_type])
            axis = random.choice(axes) if 'imu' in sensor_type else None

            fault_start_time = current_time + random.uniform(0, fault_duration * 0.2) # Inject fault a bit into the next period
            fault_end_time = fault_start_time + fault_duration

            # Generate a new normal period to inject the fault into
            fault_base_df = generate_normal_data(duration=fault_duration, sampling_rate=sampling_rate)
            fault_base_df.index = fault_base_df.index + pd.to_timedelta(current_time, unit='s')

            faulted_df = inject_fault(fault_base_df, sensor_type, fault_type, current_time, current_time + fault_duration, axis=axis)

            all_data.append(faulted_df)
            fault_label = f'{sensor_type}_{fault_type}'
            if axis is not None:
                fault_label += f'_{axis}'
            all_labels.append(pd.Series(fault_label, index=faulted_df.index))
            current_time += fault_duration

    dataset = pd.concat(all_data)
    labels = pd.concat(all_labels)

    # Ensure labels cover the entire dataset index
    labels = labels.reindex(dataset.index, fill_value='normal')

    return dataset, labels

# Example usage
# sensor_data, fault_labels = generate_dataset(num_normal_periods=3, normal_duration=30, fault_duration=10)
# print("Sensor Data Head:")
# print(sensor_data.head())
# print("\nFault Labels Head:")
# print(fault_labels.head())
# print("\nSensor Data Tail:")
# print(sensor_data.tail())
# print("\nFault Labels Tail:")
# print(fault_labels.tail())

# print("\nUnique Fault Labels:")
# print(fault_labels.unique())
```

```python
# Handle NaN values - using forward fill for time series data
features_filled = features.fillna(method='ffill')
features_filled = features_filled.fillna(method='bfill') # Handle NaNs at the beginning if any

# Instantiate and fit StandardScaler on the filled feature data
scaler = StandardScaler()
scaler.fit(features_filled)

# Transform the feature data
features_scaled = scaler.transform(features_filled)

print("NaN values handled and features scaled.")
print("Shape of scaled features:", features_scaled.shape)
```

**Reasoning**:
Reshape the scaled feature data into sequences and align the labels with these sequences for a sequence-to-one approach, then convert labels to numerical format using LabelEncoder and split into training and testing sets.



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Define sequence length
sequence_length = 10 # Example sequence length

# Reshape data into sequences (sequence-to-one approach for simplicity)
X_sequences = []
y_labels = []

# Ensure features_scaled and labels have the same number of samples
if features_scaled.shape[0] != labels.shape[0]:
    raise ValueError("Number of samples in features and labels do not match.")

for i in range(len(features_scaled) - sequence_length + 1):
    seq = features_scaled[i:i+sequence_length]
    label = labels.iloc[i + sequence_length - 1] # Label is the fault status at the end of the sequence

    # Check for NaN values within the sequence after filling and scaling - should ideally not happen
    if np.isnan(seq).any():
         print(f"Warning: NaN found in sequence starting at index {i}. Skipping sequence.")
         continue

    X_sequences.append(seq)
    y_labels.append(label)

X_sequences = np.array(X_sequences)
y_labels = np.array(y_labels)

print(f"Data reshaped into sequences with length {sequence_length}.")
print("Shape of X_sequences:", X_sequences.shape)
print("Shape of y_labels:", y_labels.shape)


# Convert labels to numerical format
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y_labels)

print("\nLabels converted to numerical format.")
print("Shape of y_encoded:", y_encoded.shape)
print("Original labels:", label_encoder.classes_)
print("Encoded labels (first 10):", y_encoded[:10])


# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_sequences, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

print("\nData split into training and testing sets.")
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
import random
import time
import statistics
import math
from collections import deque
import numpy as np
import tensorflow as tf
import joblib # To load scaler and label encoder
import pandas as pd # Import pandas for DataFrame operations

# Assume the previously defined functions (simulate_imu_reading, simulate_gps_reading,
# simulate_airspeed_reading, check_rate_of_change, check_statistical_anomaly,
# check_consistency, isolate_fault, handle_fault, exponential_smoothing,
# generate_dataset, inject_fault, generate_normal_data) are available in the environment.

# Assume the trained model, scaler, and label_encoder are saved and can be loaded.
# For this simulation, we'll assume they are available from previous steps in the notebook.
# In a real scenario, you would load them from files like this:
# loaded_model = tf.keras.models.load_model('path/to/your/model.h5')
# loaded_scaler = joblib.load('path/to/your/scaler.pkl')
# loaded_label_encoder = joblib.load('path/to/your/label_encoder.pkl')

# For the purpose of this interactive notebook, we assume the objects `model`, `scaler`, and `label_encoder`
# are available from the previous subtask executions.

# Define sequence length (must match the length used during training)
sequence_length = 10 # This should match the sequence_length used in the preprocessing step

def fault_detection_loop_with_dl(duration=30, sampling_rate=5, alpha=0.4):
    """Simulates the main loop of a fault detection system with multiple sensors,
       advanced detection methods, predictive analytics (exponential smoothing),
       and deep learning-based fault detection."""

    # Initialize deques for raw historical data for DL model input
    # These store dictionaries of raw sensor readings for each time step
    imu_accel_history_raw = deque(maxlen=sequence_length)
    imu_gyro_history_raw = deque(maxlen=sequence_length)
    gps_history_raw = deque(maxlen=sequence_length) # Store lat and lon together
    airspeed_history_raw = deque(maxlen=sequence_length)

    # Initialize deques for scalar historical data for statistical/smoothing checks
    # These store single scalar values or None
    imu_accel_x_history_scalar = deque(maxlen=10) # Keep previous size for compatibility with old checks
    imu_gyro_x_history_scalar = deque(maxlen=10)
    gps_lat_history_scalar = deque(maxlen=10)
    gps_lon_history_scalar = deque(maxlen=10) # Keep lon history though not used in current scalar checks
    airspeed_history_scalar = deque(maxlen=10)


    # Define prediction error thresholds (tune these based on expected sensor noise and behavior)
    # These are for the exponential smoothing predictive checks
    pred_error_thresholds = {
        "imu_accel": 0.5,  # Example threshold for IMU accelerometer (x-axis)
        "imu_gyro": 0.1,   # Example threshold for IMU gyroscope (x-axis)
        "gps_lat": 0.0005, # Example threshold for GPS latitude
        "airspeed": 1.0    # Example threshold for Airspeed
    }

    start_time = time.time()
    end_time = start_time + duration
    time_step = 1.0 / sampling_rate

    print("\nStarting fault detection simulation with Deep Learning...")

    while time.time() < end_time:
        current_time_sim = time.time() - start_time

        sensor_readings = {}
        detection_results = {}
        historical_data_scalar = { # Data for statistical/smoothing checks
            "imu_accel": list(imu_accel_x_history_scalar),
            "imu_gyro": list(imu_gyro_x_history_scalar),
            "gps_lat": list(gps_lat_history_scalar),
            "gps_lon": list(gps_lon_history_scalar),
            "airspeed": list(airspeed_history_scalar)
        }


        # Simulate sensor readings
        accel, gyro, imu_fault = simulate_imu_reading()
        sensor_readings["imu_accel"] = accel
        sensor_readings["imu_gyro"] = gyro

        # Pass last known GPS coordinates to simulate drift from that point
        last_lat = gps_history_raw[-1]['lat'] if gps_history_raw and gps_history_raw[-1] and gps_history_raw[-1]['lat'] is not None else None
        last_lon = gps_history_raw[-1]['lon'] if gps_history_raw and gps_history_raw[-1] and gps_history_raw[-1]['lon'] is not None else None
        lat, lon, gps_fault = simulate_gps_reading(current_lat=last_lat, current_lon=last_lon)
        sensor_readings["gps_lat"] = lat
        sensor_readings["gps_lon"] = lon


        airspeed, airspeed_fault = simulate_airspeed_reading()
        sensor_readings["airspeed"] = airspeed

        # Append raw readings to deques for DL model
        # Store as dictionaries to keep original sensor structure
        imu_accel_history_raw.append({"accel_x": accel[0] if accel is not None else None,
                                      "accel_y": accel[1] if accel is not None else None,
                                      "accel_z": accel[2] if accel is not None else None})
        imu_gyro_history_raw.append({"gyro_x": gyro[0] if gyro is not None else None,
                                     "gyro_y": gyro[1] if gyro is not None else None,
                                     "gyro_z": gyro[2] if gyro is not None else None})
        gps_history_raw.append({"lat": lat, "lon": lon})
        airspeed_history_raw.append({"airspeed": airspeed})

        # Append scalar readings for statistical/smoothing checks
        imu_accel_x_history_scalar.append(accel[0] if accel is not None else None)
        imu_gyro_x_history_scalar.append(gyro[0] if gyro is not None else None)
        gps_lat_history_scalar.append(lat)
        gps_lon_history_scalar.append(lon)
        airspeed_history_scalar.append(airspeed)


        # --- Deep Learning Fault Detection ---
        dl_prediction_result = {"dl_predicted_fault": "Cannot predict (not enough data)"}

        # Check if deques are full enough to form a sequence
        if (len(imu_accel_history_raw) == sequence_length and
            len(imu_gyro_history_raw) == sequence_length and
            len(gps_history_raw) == sequence_length and
            len(airspeed_history_raw) == sequence_length):

            # Create the sequence for the DL model
            # The order of features must match the training data: accel_x, accel_y, accel_z, gyro_x, gyro_y, gyro_z, gps_lat, gps_lon, airspeed
            current_sequence_list = []
            for i in range(sequence_length):
                step_data = []
                accel_step = imu_accel_history_raw[i]
                gyro_step = imu_gyro_history_raw[i]
                gps_step = gps_history_raw[i]
                airspeed_step = airspeed_history_raw[i]

                step_data.extend([accel_step["accel_x"], accel_step["accel_y"], accel_step["accel_z"]])
                step_data.extend([gyro_step["gyro_x"], gyro_step["gyro_y"], gyro_step["gyro_z"]])
                step_data.extend([gps_step["lat"], gps_step["lon"]])
                step_data.append(airspeed_step["airspeed"])

                current_sequence_list.append(step_data)

            current_sequence_np = np.array(current_sequence_list)

            # Handle potential None values in the sequence before scaling/prediction
            # Simple approach: interpolate missing values within the sequence
            # In a real system, more robust strategies might be needed (e.g., state estimation)
            current_sequence_df = pd.DataFrame(current_sequence_np).astype(float) # Cast to float

            # Use bfill and ffill directly to avoid FutureWarning
            current_sequence_df = current_sequence_df.interpolate(method='linear', limit_direction='both').bfill().ffill()


            # Check if any NaNs still exist after interpolation/fill (e.g., all values were None)
            if current_sequence_df.isnull().values.any():
                 dl_prediction_result = {"dl_predicted_fault": "Cannot predict (missing data after fill)"}
            else:
                current_sequence_filled_np = current_sequence_df.values

                # Scale the sequence
                # The scaler expects a 2D array (samples, features), but we have (sequence_length, features)
                # Scale each time step independently using the fitted scaler
                # Ignore UserWarning from scaler as feature names are not used here
                import warnings
                with warnings.catch_warnings():
                    warnings.simplefilter("ignore")
                    scaled_sequence = scaler.transform(current_sequence_filled_np)


                # Reshape for the DL model: (1, sequence_length, num_features)
                scaled_sequence = np.expand_dims(scaled_sequence, axis=0)

                # Make prediction
                y_pred_probs = model.predict(scaled_sequence, verbose=0)
                predicted_class_index = np.argmax(y_pred_probs, axis=1)[0]

                # Decode the prediction
                predicted_label = label_encoder.inverse_transform([predicted_class_index])[0]
                dl_prediction_result = {"dl_predicted_fault": predicted_label}

        detection_results.update(dl_prediction_result) # Add DL result to detection results


        # --- Other Fault Detection Methods (using scalar histories) ---
        # IMU Accelerometer (x-axis)
        if accel is not None:
            accel_x = accel[0]
            detection_results["imu_accel_rate_of_change"] = check_rate_of_change(accel_x, list(imu_accel_x_history_scalar))
            detection_results["imu_accel_statistical_anomaly"] = check_statistical_anomaly(accel_x, imu_accel_x_history_scalar)

            # Predictive Anomaly Check for IMU Accel (Exponential Smoothing)
            smoothed_accel = exponential_smoothing(list(imu_accel_x_history_scalar), alpha)
            if smoothed_accel and len(smoothed_accel) > 0:
                prediction = smoothed_accel[-1]
                prediction_error = abs(accel_x - prediction)
                detection_results["imu_accel_exp_smooth_predictive_anomaly"] = prediction_error > pred_error_thresholds["imu_accel"]
            else:
                 detection_results["imu_accel_exp_smooth_predictive_anomaly"] = False

        else:
            detection_results["imu_accel_rate_of_change"] = False
            detection_results["imu_accel_statistical_anomaly"] = False
            detection_results["imu_accel_exp_smooth_predictive_anomaly"] = False


        # IMU Gyroscope (x-axis)
        if gyro is not None:
            gyro_x = gyro[0]
            detection_results["imu_gyro_rate_of_change"] = check_rate_of_change(gyro_x, list(imu_gyro_x_history_scalar))
            detection_results["imu_gyro_statistical_anomaly"] = check_statistical_anomaly(gyro_x, imu_gyro_x_history_scalar)

            # Predictive Anomaly Check for IMU Gyro (Exponential Smoothing)
            smoothed_gyro = exponential_smoothing(list(imu_gyro_x_history_scalar), alpha)
            if smoothed_gyro and len(smoothed_gyro) > 0:
                prediction = smoothed_gyro[-1]
                prediction_error = abs(gyro_x - prediction)
                detection_results["imu_gyro_exp_smooth_predictive_anomaly"] = prediction_error > pred_error_thresholds["imu_gyro"]
            else:
                 detection_results["imu_gyro_exp_smooth_predictive_anomaly"] = False
        else:
            detection_results["imu_gyro_rate_of_change"] = False
            detection_results["imu_gyro_statistical_anomaly"] = False
            detection_results["imu_gyro_exp_smooth_predictive_anomaly"] = False


        # GPS Latitude
        if lat is not None:
            detection_results["gps_rate_of_change"] = check_rate_of_change(lat, list(gps_lat_history_scalar))
            detection_results["gps_statistical_anomaly"] = check_statistical_anomaly(lat, gps_lat_history_scalar)

            # Predictive Anomaly Check for GPS Lat (Exponential Smoothing)
            smoothed_gps_lat = exponential_smoothing(list(gps_lat_history_scalar), alpha)
            if smoothed_gps_lat and len(smoothed_gps_lat) > 0:
                prediction = smoothed_gps_lat[-1]
                prediction_error = abs(lat - prediction)
                detection_results["gps_exp_smooth_predictive_anomaly"] = prediction_error > pred_error_thresholds["gps_lat"]
            else:
                 detection_results["gps_exp_smooth_predictive_anomaly"] = False
        else:
            detection_results["gps_rate_of_change"] = False
            detection_results["gps_statistical_anomaly"] = False
            detection_results["gps_exp_smooth_predictive_anomaly"] = False


        # Airspeed
        if airspeed is not None:
            detection_results["airspeed_rate_of_change"] = check_rate_of_change(airspeed, list(airspeed_history_scalar))
            detection_results["airspeed_statistical_anomaly"] = check_statistical_anomaly(airspeed, airspeed_history_scalar)

            # Predictive Anomaly Check for Airspeed (Exponential Smoothing)
            smoothed_airspeed = exponential_smoothing(list(airspeed_history_scalar), alpha)
            if smoothed_airspeed and len(smoothed_airspeed) > 0:
                prediction = smoothed_airspeed[-1]
                prediction_error = abs(airspeed - prediction)
                detection_results["airspeed_exp_smooth_predictive_anomaly"] = prediction_error > pred_error_thresholds["airspeed"]
            else:
                 detection_results["airspeed_exp_smooth_predictive_anomaly"] = False
        else:
            detection_results["airspeed_rate_of_change"] = False
            detection_results["airspeed_statistical_anomaly"] = False
            detection_results["airspeed_exp_smooth_predictive_anomaly"] = False


        # Example consistency check (replace with more appropriate checks)
        if lat is not None and airspeed is not None: # Example: Check consistency between lat and airspeed
             detection_results["gps_airspeed_consistency"] = check_consistency(lat, airspeed, threshold=50) # Increased threshold
        else:
             detection_results["gps_airspeed_consistency"] = False


        # Isolate fault (updated to consider DL prediction)
        isolated_fault_info = isolate_fault_with_dl(sensor_readings, detection_results, historical_data_scalar)

        # Handle fault (updated to consider DL prediction)
        handle_fault_with_dl(isolated_fault_info)

        time.sleep(time_step)  # Control sampling rate

    print("Fault detection simulation with Deep Learning finished.")

# Update isolate_fault function to consider DL prediction
def isolate_fault_with_dl(sensor_readings, detection_results, historical_data_scalar):
  """
  Attempts to identify the likely faulty sensor based on detection results,
  including the Deep Learning prediction.

  Args:
    sensor_readings: Dictionary of current sensor readings.
    detection_results: Dictionary of detection results for each sensor/check, including 'dl_predicted_fault'.
    historical_data_scalar: Dictionary containing historical scalar data.

  Returns:
    A string indicating the likely faulty sensor and anomaly type, or "No isolated fault".
  """
  dl_predicted_fault = detection_results.get("dl_predicted_fault", "Cannot predict")

  # If DL model predicts a specific fault (and not 'normal' or 'Cannot predict'), prioritize it
  if dl_predicted_fault not in ["normal", "Cannot predict", "Cannot predict (not enough data)", "Cannot predict (missing data after fill)"]:
      return f"Likely faulty sensor (DL Prediction): {dl_predicted_fault}"

  # Otherwise, fall back to rule-based isolation from other checks
  alerting_sensors = {}
  for key, is_alerting in detection_results.items():
    if key.startswith("dl_predicted_fault"): # Skip the DL prediction itself in this loop
        continue

    if is_alerting:
      # Simple logic to associate alert with a sensor
      if "imu_accel" in key:
          sensor_name = "IMU Accelerometer"
      elif "imu_gyro" in key:
          sensor_name = "IMU Gyroscope"
      elif "gps" in key:
          sensor_name = "GPS"
      elif "airspeed" in key:
          sensor_name = "Airspeed"
      else:
          sensor_name = "Unknown Sensor" # Fallback

      if sensor_name not in alerting_sensors:
          alerting_sensors[sensor_name] = []
      alerting_sensors[sensor_name].append(key) # Store the specific check that triggered the alert

  if not alerting_sensors:
      # If DL didn't predict a fault and no other checks alerted
      return "No isolated fault"


  # Prioritize faults based on specific anomaly types or combinations from other checks
  # Prioritize exponential smoothing predictive anomalies
  for sensor_name, alerts in alerting_sensors.items():
      if any("exp_smooth_predictive_anomaly" in alert_type for alert_type in alerts):
          anomalies = ", ".join(alerts)
          return f"Likely faulty sensor: {sensor_name} (Exponential Smoothing Predictive Anomaly: {anomalies})"


  # If GPS and IMU both alert, and GPS has a sudden jump anomaly, prioritize GPS.
  if "GPS" in alerting_sensors and any("sudden_jump" in alert_type for alert_type in alerting_sensors["GPS"]):
       return "Likely faulty sensor: GPS (Anomaly: Sudden Jump detected)"

  # If IMU is alerting and has a rate of change anomaly, prioritize IMU (example)
  if "IMU Accelerometer" in alerting_sensors and any("rate_of_change" in alert_type for alert_type in alerting_sensors["IMU Accelerometer"]):
       return "Likely faulty sensor: IMU Accelerometer (Anomaly: High Rate of Change detected)"

  if "IMU Gyroscope" in alerting_sensors and any("rate_of_change" in alert_type for alert_type in alerting_sensors["IMU Gyroscope"]):
       return "Likely faulty sensor: IMU Gyroscope (Anomaly: High Rate of Change detected)"


  # Basic consistency check failure logic (example)
  if any("consistency" in alert_type for sensor_alerts in alerting_sensors.values() for alert_type in sensor_alerts):
      involved_sensors = [s for s, alerts in alerting_sensors.items() if any("consistency" in alert for alert in alerts)]
      if len(involved_sensors) > 1:
          return f"Potential consistency issue detected involving: {', '.join(involved_sensors)}. Cannot definitively isolate fault."
      else:
           return f"Likely faulty sensor: {involved_sensors[0]} (Anomaly: Consistency check failed)"


  # If no specific rule matches, indicate the sensor with the most alerts or a general message
  first_alerting_sensor = list(alerting_sensors.keys())[0]
  anomalies = ", ".join(alerting_sensors[first_alerting_sensor])
  return f"Likely faulty sensor: {first_alerting_sensor} (Multiple Anomalies: {anomalies})"


# Update handle_fault function to consider DL prediction
def handle_fault_with_dl(isolated_fault_info):
  """
  Handles a detected and isolated fault by performing a basic response,
  including specific handling for Deep Learning predictions.

  Args:
    isolated_fault_info: A string indicating the likely faulty sensor and anomaly type,
                         or "No isolated fault".
  """
  print("\n--- Handling Fault ---")
  if isolated_fault_info == "No isolated fault":
    print("No fault detected or isolated. System operating normally.")
  elif isolated_fault_info.startswith("Likely faulty sensor (DL Prediction):"):
    predicted_fault_label = isolated_fault_info.split(":")[1].strip()
    print(f"Deep Learning Model predicted fault: {predicted_fault_label}")
    # Implement specific responses based on the predicted fault label
    if predicted_fault_label == "normal":
        print("Response: DL model predicts normal operation, but other checks might be alerting. Investigate discrepancy.")
    elif "failure" in predicted_fault_label:
        print(f"Response: DL model predicts {predicted_fault_label}. Initiating emergency procedures: Attempt controlled descent or emergency landing.")
        # In a real system: Trigger critical flight control response
    elif "stuck" in predicted_fault_label:
         print(f"Response: DL model predicts {predicted_fault_label}. Switching to redundant sensor or using model-based estimation.")
         # In a real system: Switch data source, adjust control loops
    elif "noisy" in predicted_fault_label:
         print(f"Response: DL model predicts {predicted_fault_label}. Filtering sensor data more aggressively.")
         # In a real system: Apply stronger filters in estimation
    elif "sudden_jump" in predicted_fault_label:
         print(f"Response: DL model predicts {predicted_fault_label}. Discarding current reading and relying on historical data/other sensors temporarily.")
         # In a real system: Implement a data outlier rejection mechanism
    elif "loss_of_signal" in predicted_fault_label:
         print(f"Response: DL model predicts {predicted_fault_label}. Entering degraded navigation mode.")
         # In a real system: Switch navigation source
    elif "intermittent" in predicted_fault_label:
        print(f"Response: DL model predicts {predicted_fault_label}. Monitoring sensor closely, potentially switching to backup if intermittent issues persist.")
        # In a real system: Flag sensor for potential replacement, use backup if available

    else:
        print(f"Response: DL model predicted an unhandled fault type: {predicted_fault_label}. Entering a general safe mode.")

  elif "exp_smooth_predictive_anomaly" in isolated_fault_info:
    print(f"Detected potential impending issue (Exponential Smoothing): {isolated_fault_info}")
    print("Response: Logging predictive anomaly. Issuing warning to operator.")
    print("Further Action: Consider increasing sampling rate for this sensor or initiating a diagnostic check.")
    # In a real system: Log warning, potentially increase data rate, trigger diagnostics

  elif "IMU Accelerometer" in isolated_fault_info:
    print(f"Detected fault: {isolated_fault_info}")
    print("Response: Logging IMU accelerometer fault. Consider using alternative attitude estimation sources.")
    # In a real system: Log fault, potentially switch to other sensors (e.g., GPS for velocity updates)
  elif "IMU Gyroscope" in isolated_fault_info:
    print(f"Detected fault: {isolated_fault_info}")
    print("Response: Logging IMU gyroscope fault. Consider reduced maneuverability or reliance on magnetometers/GPS for heading.")
    # In a real system: Log fault, potentially reduce reliance on gyro data for control loops
  elif "GPS" in isolated_fault_info:
    print(f"Detected fault: {isolated_fault_info}")
    print("Response: Logging GPS fault. Entering degraded navigation mode (e.g., relying on IMU and airspeed). Consider returning to base or landing.")
    # In a real system: Log fault, switch navigation source, potentially trigger return-to-launch or landing sequence
  elif "Airspeed" in isolated_fault_info:
    print(f"Detected fault: {isolated_fault_info}")
    print("Response: Logging Airspeed fault. Relying on ground speed from GPS or estimated airspeed from motor performance. Exercise caution during maneuvers.")
    # In a real system: Log fault, switch airspeed source or use model-based estimation
  elif "Potential consistency issue" in isolated_fault_info:
      print(f"Detected potential issue: {isolated_fault_info}")
      print("Response: Logging potential inconsistency. Further investigation or cross-checking with other parameters is recommended.")
      # In a real system: Log warning, trigger more detailed diagnostics
  else:
    print(f"Detected an unhandled fault type: {isolated_fault_info}")
    print("Response: Logging unknown fault. Entering a general safe mode.")
    # In a real system: Log fault, enter a general safe mode with minimal functionality

  print("----------------------")


# Run the updated simulation
# Set duration and sampling rate for simulation
fault_detection_loop_with_dl(duration=60, sampling_rate=10, alpha=0.4)
```

## Implement consistency checks with thresholds and hysteresis

### Subtask:

```python
def check_consistency_with_hysteresis(reading1, reading2, current_state, warning_threshold, critical_threshold, hysteresis):
  """
  Checks consistency between two sensor readings with configurable thresholds and hysteresis.

  Args:
    reading1: The reading from the first sensor (float or None).
    reading2: The reading from the second sensor (float or None).
    current_state: The current consistency state ('normal', 'warning', 'critical', 'cannot_check').
    warning_threshold: The threshold to enter the 'warning' state from 'normal'.
    critical_threshold: The threshold to enter the 'critical' state from 'normal' or 'warning'.
    hysteresis: The value to subtract from thresholds for returning to a lower state.

  Returns:
    The new consistency state ('normal', 'warning', 'critical', 'cannot_check').
  """
  # Handle None values
  if reading1 is None or reading2 is None:
    return "cannot_check"

  # Ensure readings are numeric
  try:
      diff = abs(float(reading1) - float(reading2))
  except (ValueError, TypeError):
      print(f"Warning: Non-numeric readings provided for consistency check: {reading1}, {reading2}")
      return "cannot_check"


  if current_state == "normal":
    if diff >= critical_threshold:
      return "critical"
    elif diff >= warning_threshold:
      return "warning"
    else:
      return "normal"
  elif current_state == "warning":
    if diff >= critical_threshold:
      return "critical"
    elif diff < warning_threshold - hysteresis:
      return "normal"
    else:
      return "warning"
  elif current_state == "critical":
    if diff < critical_threshold - hysteresis:
      return "warning"
    else:
      return "critical"
  else: # current_state is "cannot_check" or an unhandled state
    # If readings are now valid, check against thresholds to re-enter a normal state
    if diff < warning_threshold - hysteresis: # Use lower threshold with hysteresis to return to normal
        return "normal"
    elif diff < critical_threshold - hysteresis:
        return "warning" # Can return to warning if below critical-hysteresis
    elif diff >= critical_threshold:
        return "critical" # Stay in critical if still above critical
    else: # Between warning-hysteresis and critical-hysteresis, or between critical-hysteresis and critical
        return current_state # Maintain the current state if within hysteresis band or between thresholds


# Example Usage (for testing)
# Initial state
# current_consistency_state = "normal"
# warning_thresh = 5.0
# critical_thresh = 10.0
# hyst = 2.0

# Test sequence of readings and states
# readings_sequence = [
#     (100, 101), # diff = 1 (normal)
#     (100, 104), # diff = 4 (normal)
#     (100, 106), # diff = 6 (warning)
#     (100, 108), # diff = 8 (warning)
#     (100, 111), # diff = 11 (critical)
#     (100, 109), # diff = 9 (critical - still above critical - hyst)
#     (100, 107), # diff = 7 (warning - below critical - hyst, above warning)
#     (100, 103), # diff = 3 (normal - below warning - hyst)
#     (100, None), # cannot_check
#     (100, 102)  # normal (if previous was cannot_check, check against threshold to transition back)
# ]

# print(f"Initial state: {current_consistency_state}")
# for r1, r2 in readings_sequence:
#     current_consistency_state = check_consistency_with_hysteresis(r1, r2, current_consistency_state, warning_thresh, critical_thresh, hyst)
#     print(f"Readings ({r1}, {r2}), New state: {current_consistency_state}")
```

## Explanation of New Anomaly Detection Methods

### ML-Based Anomaly Detection (Isolation Forest)

We've introduced an **Isolation Forest** model as an example of an ML-based anomaly detection technique. Here's how it works and why it's useful:

*   **How it Works:** Isolation Forest works by randomly selecting a feature and then randomly selecting a split value between the maximum and minimum values of that selected feature. This process recursively partitions the data. Anomalies are typically few and different from normal observations, making them easier to isolate. The number of splits required to isolate a data point is a measure of its abnormality. Data points that require fewer splits to be isolated are considered more likely to be anomalies.
*   **Training:** The model is trained on a dataset of *normal* sensor readings. It learns the characteristics of what constitutes "normal" behavior for the system based on the relationships between different sensor readings.
*   **Detection:** During the simulation, the current sensor readings (as a single data point across all features) are fed into the trained Isolation Forest model. The model outputs an anomaly score or a prediction (-1 for anomaly, 1 for normal). If the prediction is -1, it indicates that the current state of the system, based on the combined sensor readings, is significantly different from the normal behavior the model learned.
*   **Benefits:**
    *   **Efficiency:** Isolation Forest is generally efficient and scalable, even for datasets with many features.
    *   **Effective for Outliers:** It's particularly good at detecting distinct outliers.
    *   **Unsupervised:** It doesn't require labeled data for anomalies, only examples of normal behavior.
*   **Limitations:**
    *   **Defining "Normal":** Its performance depends heavily on the quality and representativeness of the "normal" data used for training. If the training data doesn't cover all normal operating conditions, it might flag legitimate variations as anomalies.
    *   **Parameter Tuning:** The `contamination` parameter (estimating the proportion of anomalies in the data) needs careful tuning.
    *   **Not Fault Classification:** It detects *anomalies* in the system state but doesn't inherently tell you *which sensor* is causing the anomaly or the *type* of fault. This still requires integration with fault isolation logic.

### Consistency Checks with Thresholds and Hysteresis

We've enhanced the consistency checks to include configurable thresholds and hysteresis. This makes them more robust to noisy data and prevents rapid toggling between states ("chattering").

*   **How it Works:**
    *   **Comparison:** It compares the readings of two related sensors (e.g., GPS latitude and airspeed in our simplified example, though a more realistic check would be GPS ground speed vs airspeed under specific conditions, or comparing redundant sensors).
    *   **Difference Calculation:** The absolute difference between the two readings is calculated.
    *   **Thresholds:** Instead of a single threshold, we use two: a `warning_threshold` and a `critical_threshold`.
    *   **Hysteresis:** Hysteresis introduces a "lag" in the state change. To transition from a higher alert state (e.g., 'warning' or 'critical') back to a lower state (e.g., 'normal' or 'warning'), the difference between the readings must drop *below* the threshold for the lower state minus the hysteresis value. This prevents the state from rapidly switching if the difference hovers around a threshold due to noise.
    *   **States:** The check maintains a state ('normal', 'warning', 'critical', 'cannot_check'). The state transitions based on the difference and the current state, incorporating the hysteresis logic.
*   **Integration:** In the simulation loop, we now perform this enhanced consistency check for the defined sensor pairs and update the `detection_results` with the current state of each check.
*   **Benefits:**
    *   **Robustness:** Hysteresis significantly reduces false alarms caused by noisy data or minor fluctuations near a threshold.
    *   **Graduated Alerts:** Using warning and critical thresholds allows for a more nuanced response to inconsistencies.
    *   **Physically Interpretable:** Consistency checks often compare physically related quantities, which can be easier to interpret than abstract anomaly scores.
*   **Limitations:**
    *   **Requires Related Sensors:** Only applicable when there are physically related or redundant sensor measurements available.
    *   **Threshold Tuning:** Requires careful tuning of warning, critical, and hysteresis thresholds based on sensor specifications and expected operational variations.
    *   **Doesn't Isolate Fault:** Like basic anomaly detection, a consistency check failure indicates an issue *between* two sensors, not necessarily which one is faulty. Fault isolation is still needed.
    *   **Handling Complex Relationships:** For sensors with non-linear or dynamic relationships, simple difference checks may not be sufficient; more advanced model-based or ML-based consistency checks might be needed.

These new methods provide a more robust and potentially more sensitive approach to detecting anomalies and inconsistencies compared to the initial basic checks.

## Explanation of Risk Matrix and Fault Classification

We've implemented a risk matrix and fault classification system to prioritize detected issues based on their potential severity. This is a common practice in safety-critical systems to ensure that the most dangerous faults receive immediate attention and appropriate responses.

### Risk Levels

We defined several risk levels, each associated with a numerical value for internal use and a descriptive term:

*   **Normal (0):** No significant risk detected.
*   **Low (1):** Minor impact, unlikely to occur frequently.
*   **Medium (2):** Moderate impact or higher likelihood of a minor impact fault.
*   **High (3):** Significant impact or likely to occur frequently.
*   **Critical (4):** Catastrophic impact, regardless of likelihood, or highly likely major impact.

These levels provide a structured way to think about the potential consequences of different faults.

### Likelihood and Impact Assessment

To determine the risk level, we estimated the **likelihood** and **impact** of each specific fault type simulated in our system:

*   **Likelihood:** An estimation of how often a particular fault is expected to occur (e.g., Rare, Unlikely, Possible, Likely, Frequent). These are based on hypothetical assumptions about sensor reliability in our simulation. In a real system, these would be informed by historical data, manufacturer specifications, and operational experience.
*   **Impact:** An assessment of the potential consequences of a fault on the VTOL's mission, safety, and functionality (e.g., Negligible, Minor, Moderate, Major, Catastrophic). These are also hypothetical in our simulation but would be based on detailed system analysis in a real-world scenario.

### The Risk Matrix

The core of the risk classification is the risk matrix, which combines the likelihood and impact assessments to determine the overall risk level:

| Likelihood \ Impact | Negligible | Minor    | Moderate | Major      | Catastrophic |
| :------------------ | :--------- | :------- | :------- | :--------- | :----------- |
| **Rare**            | Normal     | Low      | Low      | Medium     | High         |
| **Unlikely**        | Low        | Low      | Medium   | High       | High         |
| **Possible**        | Low        | Medium   | High     | High       | Critical     |
| **Likely**          | Medium     | High     | High     | Critical   | Critical     |
| **Frequent**        | High       | High     | Critical | Critical   | Critical     |

For example, a fault with "Possible" likelihood and "Major" impact is classified as "Critical" risk.
