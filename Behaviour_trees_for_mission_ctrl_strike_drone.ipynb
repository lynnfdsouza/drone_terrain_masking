## Develop Selector and Sequence Nodes

### Subtask:
Create selector and sequence nodes to manage the execution of different tasks and behaviors based on the mission requirements and environmental conditions.

**Reasoning**:
Selector nodes are used when the drone needs to choose among several possible behaviors (e.g., which mission to execute). They tick their children in order and succeed as soon as one child succeeds. If all children fail, the selector fails. Sequence nodes are used for tasks that require a specific order of actions (e.g., identify target, lock on, fire missile). They tick their children in order and succeed only if all children succeed. If any child fails, the sequence fails. We will define conceptual structures for these nodes.

```python
# Conceptual design of Selector and Sequence nodes

# Selector Node:
# - Evaluates child nodes in order.
# - Returns SUCCESS if any child returns SUCCESS.
# - Returns RUNNING if any child returns RUNNING.
# - Returns FAILURE if all children return FAILURE.
class Selector:
    def __init__(self, children):
        self.children = children

    def tick(self):
        for child in self.children:
            status = child.tick()
            if status == 'SUCCESS' or status == 'RUNNING':
                return status
        return 'FAILURE'

# Sequence Node:
# - Evaluates child nodes in order.
# - Returns SUCCESS if all children return SUCCESS.
# - Returns RUNNING if any child returns RUNNING.
# - Returns FAILURE if any child returns FAILURE.
class Sequence:
    def __init__(self, children):
        self.children = children

    def tick(self):
        for child in self.children:
            status = child.tick()
            if status == 'FAILURE' or status == 'RUNNING':
                return status
        return 'SUCCESS'

print("Conceptual design of Selector and Sequence nodes completed.")
```

**Reasoning**:
Leaf nodes are the terminal nodes in a behaviour tree. They perform the actual work, such as executing an action or checking a condition. Unlike composite nodes (Selectors and Sequences), leaf nodes do not have children. They return a status of SUCCESS, FAILURE, or RUNNING based on the outcome of their action or condition check. The error `IndexError: list index out of range` occurs because the node logs (`engage_target_sequence.log` and `patrol_node.log`) are being cleared at the end of *each* simulation step within the `collect_logs` function. When the analysis code tries to access `log[-1]`, the list is empty. The logging mechanism needs to be adjusted to append to a central list for the scenario within the nodes themselves, or the logs should be collected differently without clearing them immediately. A better approach is to let the nodes append to their own logs, and then collect all logs at the end of the simulation for that scenario. Also, the `sim_env` status at the end of the loop is the correct indicator, not necessarily the last logged status of individual nodes. The analysis should primarily use the collected `df_ticks` DataFrame.



```python
# Modify node classes to log directly to their own log list
class Selector:
    def __init__(self, name, children, env):
        self.name = name
        self.children = children
        self.env = env
        self.log = [] # List to store tick data

    def tick(self):
        step = self.env.time
        # print(f"Ticking Selector: {self.name} at step {step}") # Keep print for debugging/trace

        status_to_log = 'RUNNING' # Initial status before children are ticked
        self.log.append({"node": self.name, "type": "Selector", "step": step, "status": status_to_log, "action": None})

        final_status = 'FAILURE' # Default if no child succeeds or runs
        for child in self.children:
            status = child.tick()
            if status == 'SUCCESS' or status == 'RUNNING':
                final_status = status
                break # Selector stops after first success or running

        # Update the status of the selector's tick in the log entry for the current step
        # Find the log entry for the current step and update its status
        for entry in reversed(self.log): # Look from the end for the most recent entry
            if entry['step'] == step and entry['node'] == self.name:
                entry['status'] = final_status
                break

        return final_status


class Sequence:
    def __init__(self, name, children, env):
        self.name = name
        self.children = children
        self.env = env
        self.log = [] # List to store tick data

    def tick(self):
        step = self.env.time
        # print(f"Ticking Sequence: {self.name} at step {step}") # Keep print for debugging/trace

        status_to_log = 'RUNNING' # Initial status before children are ticked
        self.log.append({"node": self.name, "type": "Sequence", "step": step, "status": status_to_log, "action": None})

        final_status = 'SUCCESS' # Default if all children succeed
        for child in self.children:
            status = child.tick()
            if status == 'FAILURE' or status == 'RUNNING':
                final_status = status
                break # Sequence stops after first failure or running

        # Update the status of the sequence's tick in the log entry for the current step
        # Find the log entry for the current step and update its status
        for entry in reversed(self.log): # Look from the end for the most recent entry
            if entry['step'] == step and entry['node'] == self.name:
                entry['status'] = final_status
                break

        return final_status

class Leaf:
    def __init__(self, name, env):
        self.name = name
        self.env = env
        self.log = [] # List to store tick data

    def tick(self):
        step = self.env.time
        # print(f"Ticking Leaf: {self.name} at step {step}") # Keep print for debugging/trace
        status, action_taken = self._execute() # Modify _execute to return status and action

        # Log the tick with status and action
        self.log.append({"node": self.name, "type": self.__class__.__name__, "step": step, "status": status, "action": action_taken})

        return status

    def _execute(self):
        """Subclasses must implement and return (status, action_taken)."""
        raise NotImplementedError("Subclasses must implement abstract method")

# Modified Leaf Node: Patrol Area
class PatrolArea(Leaf):
    def __init__(self, name, env):
        super().__init__(name, env)
        self.patrol_direction = "east" # Simple patrol logic

    def _execute(self):
        print(f"Drone is patrolling the designated area (Node: {self.name}).")
        action = None
        # Simple patrol: move in one direction, then switch when hitting boundary
        if self.patrol_direction == "east":
            if self.env.drone_pos[0] < self.env.area_size - 10:
                action = "move_east"
                self.env.update_state(action)
            else:
                self.patrol_direction = "south"
                action = "move_south"
                self.env.update_state(action) # Move immediately after changing direction
        elif self.patrol_direction == "south":
             if self.env.drone_pos[1] > 10:
                action = "move_south"
                self.env.update_state(action)
             else:
                self.patrol_direction = "west"
                action = "move_west"
                self.env.update_state(action)
        elif self.patrol_direction == "west":
             if self.env.drone_pos[0] > 10:
                action = "move_west"
                self.env.update_state(action)
             else:
                self.patrol_direction = "north"
                action = "move_north"
                self.env.update_state(action)
        elif self.patrol_direction == "north":
             if self.env.drone_pos[1] < self.env.area_size - 10:
                action = "move_north"
                self.env.update_state(action)
             else:
                self.patrol_direction = "east"
                action = "move_east"
                self.env.update_state(action)


        # Patrol is an ongoing action, so it typically returns RUNNING
        return 'RUNNING', action

# Modified Leaf Node: Identify Target
class IdentifyTarget(Leaf):
    def __init__(self, name, env):
        super().__init__(name, env)

    def _execute(self):
        action = None # Identify is a condition check, no direct action on environment state (except setting current_target)
        if self.env.identify_target_in_range():
            print(f"Identify Target returned SUCCESS (Node: {self.name}).")
            return 'SUCCESS', action
        else:
            print(f"Identify Target returned FAILURE (Node: {self.name}).")
            return 'FAILURE', action

# Modified Leaf Node: Fire Missile
class FireMissile(Leaf):
    def __init__(self, name, env):
        super().__init__(name, env)

    def _execute(self):
        action = "fire_missile" # The action is firing the missile
        # Fire missile only if a target was identified in the previous step and is active
        if hasattr(self.env, 'current_target') and self.env.current_target is not None and self.env.targets[self.env.current_target]['status'] == 'Active':
             self.env.update_state(action)
             # Assuming firing is quick and successful if target active
             print(f"Fire Missile returned SUCCESS (Node: {self.name}).")
             return 'SUCCESS', action
        else:
             print(f"Fire Missile returned FAILURE (no active target) (Node: {self.name}).")
             return 'FAILURE', None # No action taken if fire fails


# Modified Leaf Node: Check Fuel Level
class CheckFuelLevel(Leaf):
    def __init__(self, name, env):
        super().__init__(name, env)

    def _execute(self):
        action = None # Checking fuel is a condition check
        if self.env.check_fuel_level():
            print(f"Check Fuel Level returned SUCCESS (Node: {self.name}).")
            return 'SUCCESS', action
        else:
            print(f"Check Fuel Level returned FAILURE (low fuel) (Node: {self.name}).")
            return 'FAILURE', action


# Define the list of scenarios to run
scenarios_to_run = [scenario_1, scenario_2, scenario_3]

# Cumulative dictionary to store all tick data across all scenarios
all_scenario_tick_data = {}


# Iterate through the scenarios
for scenario in scenarios_to_run:
    print(f"\n--- Running Scenario: {scenario['name']} ---")
    print(f"Description: {scenario['description']}")

    # Create a new instance of the DroneSimulationEnvironment for the current scenario
    sim_env = DroneSimulationEnvironment(
        area_size=scenario["params"]["area_size"],
        initial_fuel=scenario["params"]["initial_fuel"]
    )
    if "targets" in scenario["params"]:
         sim_env.targets = scenario["params"]["targets"].copy() # Use copy to avoid modifying the original scenario dict


    sim_env.drone_pos = scenario["initial_drone_pos"].copy() # Use copy

    # Create instances of the modified leaf nodes, passing the new environment instance and a name
    patrol_node = PatrolArea("PatrolArea", sim_env)
    identify_target_node = IdentifyTarget("IdentifyTarget", sim_env)
    fire_missile_node = FireMissile("FireMissile", sim_env)
    check_fuel_node = CheckFuelLevel("CheckFuelLevel", sim_env)

    # Recreate the engage_target_sequence and mission_selector nodes with names and the environment
    engage_target_sequence = Sequence("EngageTargetSequence", [identify_target_node, check_fuel_node, fire_missile_node], sim_env)
    mission_selector = Selector("MissionSelector", [engage_target_sequence, patrol_node], sim_env)

    # Set the root_node for the behaviour tree
    root_node = mission_selector

    # Implement a simulation loop for the current scenario
    step = 0
    max_simulation_steps = 500

    print("\n--- Starting Simulation Loop for Scenario ---")
    while sim_env.status == "Operating" and step < max_simulation_steps:
        step += 1
        # print(f"\n--- Simulation Step {step} ---") # Uncomment for more detailed step-by-step output

        # Tick the root node
        tree_status = root_node.tick()

        # No need to collect logs within the loop. Logs are appended to node.log lists.

        # print(f"Behaviour tree root returned status: {tree_status}") # Uncomment for more detailed step-by-step output

    # After the simulation loop finishes, collect all logs for the scenario
    scenario_tick_data = []
    def collect_all_logs(node, log_list):
        log_list.extend(node.log)
        if hasattr(node, 'children'):
            for child in node.children:
                collect_all_logs(child, log_list)

    collect_all_logs(root_node, scenario_tick_data)

    # Store the collected data for the current scenario
    all_scenario_tick_data[scenario['name']] = scenario_tick_data

    # After the simulation loop finishes, print a summary
    print("\n--- Simulation Loop Finished for Scenario ---")
    print(f"Scenario: {scenario['name']}")
    print(f"Final Simulation Status: {sim_env.status}")
    print(f"Total Simulation Steps: {step}")
    print(f"Remaining Fuel: {sim_env.drone_fuel:.2f}")
    target_statuses_list = [f'{tid}: {t["status"]}' for tid, t in sim_env.targets.items()]
    print(f"Targets Status: {target_statuses_list}")

    print(f"\n--- Finished Scenario: {scenario['name']} ---\n")

print("Data collection during simulation completed.")

# Analysis of collected data
print("\n--- Analyzing Simulation Data ---")

# Iterate through the collected data for each scenario
for scenario_name, tick_data in all_scenario_tick_data.items():
    print(f"\n--- Analysis for Scenario: {scenario_name} ---")
    print(f"Total ticks recorded: {len(tick_data)}")

    if not tick_data:
        print("No tick data recorded for this scenario.")
        continue

    # Convert tick data to a pandas DataFrame for easier analysis
    df_ticks = pd.DataFrame(tick_data)
    # print("\nTick Data DataFrame (first 10 rows):")
    # display(df_ticks.head(10)) # Use display in notebook

    # Analyze the sequence of node ticks and their statuses
    print("\nExecution Trace Summary (first 20 ticks):")
    # Print limited rows to avoid overwhelming output
    display(df_ticks[['step', 'node', 'type', 'status', 'action']].head(20))

    # Analyze the outcome of key nodes
    print("\nSummary of key node outcomes:")
    # For example, check how many times IdentifyTarget succeeded or failed
    identify_target_outcomes = df_ticks[df_ticks['node'] == 'IdentifyTarget']['status'].value_counts()
    print(f"IdentifyTarget outcomes:\n{identify_target_outcomes}")

    # Check FireMissile outcomes
    fire_missile_outcomes = df_ticks[df_ticks['node'] == 'FireMissile']['status'].value_counts()
    print(f"FireMissile outcomes:\n{fire_missile_outcomes}")

    # Check CheckFuelLevel outcomes
    check_fuel_outcomes = df_ticks[df_ticks['node'] == 'CheckFuelLevel']['status'].value_counts()
    print(f"CheckFuelLevel outcomes:\n{check_fuel_outcomes}")

    # Analyze the overall mission outcome based on the last recorded state or log
    print("\nOverall Scenario Outcome:")

    # Find the last recorded step to see what happened at the end
    last_step_data = df_ticks[df_ticks['step'] == df_ticks['step'].max()]
    print(f"Behavior at last step ({df_ticks['step'].max()}):")
    display(last_step_data[['node', 'type', 'status', 'action']])

    # Compare with expected outcome and summarize observations
    expected_outcome = None
    scenario_description = ""
    # Need to get the final sim_env status for the current scenario run
    # This requires accessing the sim_env from the loop scope or storing it.
    # Let's re-run the scenarios quickly to get the final sim_env status for analysis.
    # This is not ideal but works with the current structure.
    # A better structure would store the final sim_env state with the tick data.

    # Rerun scenario to get final sim_env state (temporary workaround)
    temp_sim_env = None
    for scenario in scenarios_to_run:
        if scenario['name'] == scenario_name:
            temp_sim_env = DroneSimulationEnvironment(
                area_size=scenario["params"]["area_size"],
                initial_fuel=scenario["params"]["initial_fuel"]
            )
            if "targets" in scenario["params"]:
                 temp_sim_env.targets = scenario["params"]["targets"].copy()
            temp_sim_env.drone_pos = scenario["initial_drone_pos"].copy()

            # Rebuild tree for temp run (no logging needed here)
            temp_patrol_node = PatrolArea("PatrolArea", temp_sim_env)
            temp_identify_target_node = IdentifyTarget("IdentifyTarget", temp_sim_env)
            temp_fire_missile_node = FireMissile("FireMissile", temp_sim_env)
            temp_check_fuel_node = CheckFuelLevel("CheckFuelLevel", temp_sim_env)
            temp_engage_target_sequence = Sequence("EngageTargetSequence", [temp_identify_target_node, temp_check_fuel_node, temp_fire_missile_node], temp_sim_env)
            temp_mission_selector = Selector("MissionSelector", [temp_engage_target_sequence, temp_patrol_node], temp_sim_env)
            temp_root_node = temp_mission_selector

            # Run temp simulation
            step = 0
            while temp_sim_env.status == "Operating" and step < max_simulation_steps:
                step += 1
                temp_root_node.tick() # Tick without logging side effects

            scenario_description = scenario['description']
            if scenario_name == "Successful Engagement":
                expected_outcome = "Mission Accomplished (Target Destroyed)"
            elif scenario_name == "No Target Found":
                expected_outcome = "Patrol until max steps or fuel depleted"
            elif scenario_name == "Low Fuel":
                expected_outcome = "Fuel Depleted during mission"
            break


    print(f"\nScenario Description: {scenario_description}")
    print(f"Expected Outcome: {expected_outcome}")
    print(f"Actual Final Simulation Status: {temp_sim_env.status}") # Use status from temp run


    print("\nObservations:")
    if scenario_name == "Successful Engagement":
        if temp_sim_env.status == "Mission Accomplished":
            print("- The drone successfully identified and destroyed the target.")
            print("- The tree transitioned from engagement sequence to patrolling.")
        elif temp_sim_env.status == "Operating" and 'SUCCESS' in fire_missile_outcomes.index:
             print("- The drone fired a missile successfully but the simulation ended before mission accomplished status was set (e.g., multiple targets or patrol continued).")
        elif temp_sim_env.status == "Fuel Depleted":
             print("- The drone ran out of fuel before completing the mission.")
        else:
            # Check the tick data to see why it didn't succeed or deplete fuel
            if 'FAILURE' in identify_target_outcomes.index and identify_target_outcomes['FAILURE'] > 0:
                 print("- The IdentifyTarget node consistently failed, preventing engagement.")
            elif 'FAILURE' in check_fuel_outcomes.index and check_fuel_outcomes['FAILURE'] > 0:
                 print("- The CheckFuelLevel node returned FAILURE, preventing engagement.")
            else:
                 print("- Unexpected behavior observed in Successful Engagement scenario - unable to determine exact cause from logs.")


    elif scenario_name == "No Target Found":
         if temp_sim_env.status == "Operating" and 'FAILURE' in identify_target_outcomes.index and identify_target_outcomes['FAILURE'] > 0:
             print("- The drone consistently failed to identify a target, as expected.")
             print("- The tree correctly defaulted to the PatrolArea node for the entire simulation.")
             print(f"- The simulation ran until the maximum steps ({max_simulation_steps}) were reached.")
         elif temp_sim_env.status == "Fuel Depleted":
             print("- The drone ran out of fuel while patrolling, which is a possible outcome when no targets are found and the simulation runs long enough.")
         else:
             print("- Unexpected behavior observed in No Target Found scenario.")

    elif scenario_name == "Low Fuel":
        if temp_sim_env.status == "Fuel Depleted":
            print("- The drone's fuel level dropped below the threshold, as expected.")
            print("- The CheckFuelLevel node likely returned FAILURE at some point, preventing further actions in the engagement sequence.")
            print("- The simulation ended due to fuel depletion.")
        elif temp_sim_env.status == "Mission Accomplished":
             print("- The drone completed the mission despite starting with low fuel. This indicates the mission was very short or the fuel consumption rate is low relative to the task.")
        else:
             print("- Unexpected behavior observed in Low Fuel scenario.")

    # Add checks for other potential issues like infinite loops (indicated by hitting max_simulation_steps)
    if temp_sim_env.status == "Operating" and step == max_simulation_steps:
        print(f"- Simulation reached the maximum steps ({max_simulation_steps}) without concluding the mission or depleting fuel. This might indicate an issue with mission completion conditions or patrol logic preventing a terminal state.")


print("\n--- Data Analysis Completed ---")
```

```python
import random
import time
from enum import Enum
from dataclasses import dataclass
from typing import Dict, List, Optional, Tuple

class DroneStatus(Enum):
    IDLE = "idle"
    PATROLLING = "patrolling"
    ENGAGING = "engaging"
    RETURNING = "returning"
    LANDED = "landed"
    EMERGENCY = "emergency"

class SystemStatus(Enum):
    OPERATIONAL = "operational"
    DEGRADED = "degraded"
    FAILED = "failed"

class MissionOutcome(Enum):
    SUCCESS = "success"
    FAILURE = "failure"
    ABORTED = "aborted"
    PARTIAL_SUCCESS = "partial_success"

@dataclass
class DroneState:
    """Current state of the drone"""
    position: Tuple[float, float, float]  # x, y, altitude
    fuel_level: float  # 0.0 to 1.0
    status: DroneStatus
    target_detected: bool
    target_position: Optional[Tuple[float, float, float]]
    obstacle_detected: bool
    obstacle_position: Optional[Tuple[float, float, float]]
    return_to_base_commanded: bool
    system_health: Dict[str, SystemStatus]
    mission_time: float
    max_mission_time: float
    base_position: Tuple[float, float, float]

@dataclass
class TestScenario:
    """Defines a test scenario for the behavior tree"""
    name: str
    description: str
    initial_state: DroneState
    events: List[Dict]  # Time-based events that occur during the scenario
    expected_outcome: MissionOutcome
    success_criteria: List[str]

class DroneTestScenarios:
    """Collection of test scenarios for drone mission control behavior trees"""

    def __init__(self):
        self.base_position = (0.0, 0.0, 0.0)
        self.patrol_area = [(100, 100, 50), (200, 200, 50)]  # patrol boundaries

    def create_base_drone_state(self) -> DroneState:
        """Create a basic drone state with default values"""
        return DroneState(
            position=self.base_position,
            fuel_level=1.0,
            status=DroneStatus.IDLE,
            target_detected=False,
            target_position=None,
            obstacle_detected=False,
            obstacle_position=None,
            return_to_base_commanded=False,
            system_health={
                "navigation": SystemStatus.OPERATIONAL,
                "sensors": SystemStatus.OPERATIONAL,
                "communication": SystemStatus.OPERATIONAL,
                "propulsion": SystemStatus.OPERATIONAL,
                "weapons": SystemStatus.OPERATIONAL
            },
            mission_time=0.0,
            max_mission_time=1800.0,  # 30 minutes
            base_position=self.base_position
        )

    def scenario_1_successful_engagement(self) -> TestScenario:
        """Scenario 1: Successful target identification and engagement"""
        initial_state = self.create_base_drone_state()

        events = [
            {"time": 60, "type": "start_patrol", "data": {}},
            {"time": 300, "type": "target_detected", "data": {"position": (150, 150, 45)}},
            {"time": 320, "type": "target_confirmed", "data": {"threat_level": "high"}},
            {"time": 340, "type": "engagement_successful", "data": {"target_neutralized": True}},
            {"time": 360, "type": "mission_complete", "data": {}}
        ]

        return TestScenario(
            name="Successful Target Engagement",
            description="Drone successfully patrols, identifies target, and engages",
            initial_state=initial_state,
            events=events,
            expected_outcome=MissionOutcome.SUCCESS,
            success_criteria=[
                "Target detected within patrol area",
                "Target successfully engaged",
                "Drone returns to base safely",
                "Mission completed within time limit"
            ]
        )

    def scenario_2_no_target_found(self) -> TestScenario:
        """Scenario 2: No target found in patrol area"""
        initial_state = self.create_base_drone_state()

        events = [
            {"time": 60, "type": "start_patrol", "data": {}},
            {"time": 1500, "type": "patrol_complete", "data": {"targets_found": 0}},
            {"time": 1520, "type": "return_to_base", "data": {"reason": "mission_complete"}}
        ]

        return TestScenario(
            name="No Target Found",
            description="Drone completes patrol without finding any targets",
            initial_state=initial_state,
            events=events,
            expected_outcome=MissionOutcome.SUCCESS,
            success_criteria=[
                "Complete patrol area coverage",
                "No false target identifications",
                "Safe return to base",
                "Mission logged as complete"
            ]
        )

    def scenario_3_low_fuel_emergency(self) -> TestScenario:
        """Scenario 3: Low fuel during mission"""
        initial_state = self.create_base_drone_state()
        initial_state.fuel_level = 0.6  # Start with lower fuel

        events = [
            {"time": 60, "type": "start_patrol", "data": {}},
            {"time": 200, "type": "fuel_warning", "data": {"fuel_level": 0.25}},
            {"time": 220, "type": "target_detected", "data": {"position": (180, 180, 40)}},
            {"time": 240, "type": "fuel_critical", "data": {"fuel_level": 0.15}},
            {"time": 250, "type": "emergency_return", "data": {"reason": "low_fuel"}},
            {"time": 400, "type": "emergency_landing", "data": {"fuel_remaining": 0.05}}
        ]

        return TestScenario(
            name="Low Fuel Emergency",
            description="Drone encounters low fuel situation during mission",
            initial_state=initial_state,
            events=events,
            expected_outcome=MissionOutcome.ABORTED,
            success_criteria=[
                "Low fuel detected and handled",
                "Emergency return initiated",
                "Safe landing achieved",
                "Mission aborted appropriately"
            ]
        )

    def scenario_4_obstacle_encounter(self) -> TestScenario:
        """Scenario 4: Encountering unexpected obstacle"""
        initial_state = self.create_base_drone_state()

        events = [
            {"time": 60, "type": "start_patrol", "data": {}},
            {"time": 150, "type": "obstacle_detected", "data": {"position": (120, 130, 50), "type": "building"}},
            {"time": 160, "type": "path_replanned", "data": {"new_route": True}},
            {"time": 300, "type": "target_detected", "data": {"position": (160, 170, 35)}},
            {"time": 350, "type": "obstacle_during_engagement", "data": {"position": (155, 165, 40)}},
            {"time": 370, "type": "engagement_repositioned", "data": {"new_position": (140, 160, 45)}},
            {"time": 390, "type": "engagement_successful", "data": {}}
        ]

        return TestScenario(
            name="Obstacle Encounter",
            description="Drone encounters obstacles during patrol and engagement",
            initial_state=initial_state,
            events=events,
            expected_outcome=MissionOutcome.SUCCESS,
            success_criteria=[
                "Obstacles detected and avoided",
                "Path replanning successful",
                "Mission objectives still achieved",
                "No collision incidents"
            ]
        )

    def scenario_5_return_to_base_command(self) -> TestScenario:
        """Scenario 5: Receiving command to return to base"""
        initial_state = self.create_base_drone_state()

        events = [
            {"time": 60, "type": "start_patrol", "data": {}},
            {"time": 200, "type": "target_detected", "data": {"position": (170, 140, 45)}},
            {"time": 220, "type": "target_analysis", "data": {"analyzing": True}},
            {"time": 240, "type": "return_to_base_command", "data": {"priority": "high", "reason": "mission_change"}},
            {"time": 250, "type": "engagement_aborted", "data": {"target_disengaged": True}},
            {"time": 260, "type": "returning_to_base", "data": {}},
            {"time": 400, "type": "landed_at_base", "data": {}}
        ]

        return TestScenario(
            name="Return to Base Command",
            description="Drone receives priority command to return to base during active mission",
            initial_state=initial_state,
            events=events,
            expected_outcome=MissionOutcome.ABORTED,
            success_criteria=[
                "Command received and acknowledged",
                "Current operations safely terminated",
                "Immediate return initiated",
                "Successful base landing"
            ]
        )

    def scenario_6_system_failures(self) -> TestScenario:
        """Scenario 6: Multiple system failures"""
        initial_state = self.create_base_drone_state()

        events = [
            {"time": 60, "type": "start_patrol", "data": {}},
            {"time": 180, "type": "sensor_degradation", "data": {"system": "sensors", "status": "degraded"}},
            {"time": 220, "type": "target_detected", "data": {"position": (140, 160, 50), "confidence": 0.6}},
            {"time": 300, "type": "communication_failure", "data": {"system": "communication", "status": "failed"}},
            {"time": 320, "type": "autonomous_mode", "data": {"reason": "communication_lost"}},
            {"time": 400, "type": "navigation_error", "data": {"system": "navigation", "status": "degraded"}},
            {"time": 420, "type": "emergency_protocol", "data": {"mode": "safe_return"}},
            {"time": 600, "type": "emergency_landing", "data": {"location": "safe_zone"}}
        ]

        return TestScenario(
            name="Multiple System Failures",
            description="Drone experiences cascading system failures during mission",
            initial_state=initial_state,
            events=events,
            expected_outcome=MissionOutcome.FAILURE,
            success_criteria=[
                "System failures detected",
                "Emergency protocols activated",
                "Safe landing achieved",
                "No catastrophic failure"
            ]
        )

    def scenario_7_weather_emergency(self) -> TestScenario:
        """Scenario 7: Severe weather encounter"""
        initial_state = self.create_base_drone_state()

        events = [
            {"time": 60, "type": "start_patrol", "data": {}},
            {"time": 150, "type": "weather_warning", "data": {"type": "high_winds", "severity": "moderate"}},
            {"time": 200, "type": "target_detected", "data": {"position": (165, 145, 40)}},
            {"time": 250, "type": "weather_critical", "data": {"type": "thunderstorm", "severity": "severe"}},
            {"time": 260, "type": "engagement_impossible", "data": {"reason": "weather_conditions"}},
            {"time": 270, "type": "emergency_return", "data": {"reason": "weather"}},
            {"time": 300, "type": "navigation_difficult", "data": {"visibility": "poor"}},
            {"time": 450, "type": "safe_landing", "data": {"assisted": True}}
        ]

        return TestScenario(
            name="Weather Emergency",
            description="Drone encounters severe weather during mission",
            initial_state=initial_state,
            events=events,
            expected_outcome=MissionOutcome.ABORTED,
            success_criteria=[
                "Weather conditions recognized",
                "Mission safety prioritized",
                "Safe return navigation",
                "Successful emergency landing"
            ]
        )

    def scenario_8_multiple_targets(self) -> TestScenario:
        """Scenario 8: Multiple targets with priority management"""
        initial_state = self.create_base_drone_state()

        events = [
            {"time": 60, "type": "start_patrol", "data": {}},
            {"time": 180, "type": "target_detected", "data": {"position": (140, 150, 45), "priority": "medium", "id": 1}},
            {"time": 200, "type": "target_detected", "data": {"position": (180, 170, 40), "priority": "high", "id": 2}},
            {"time": 220, "type": "target_prioritized", "data": {"selected_target": 2, "reason": "higher_priority"}},
            {"time": 240, "type": "engagement_started", "data": {"target_id": 2}},
            {"time": 280, "type": "engagement_successful", "data": {"target_id": 2}},
            {"time": 300, "type": "target_reevaluated", "data": {"target_id": 1, "still_present": True}},
            {"time": 320, "type": "engagement_started", "data": {"target_id": 1}},
            {"time": 360, "type": "engagement_successful", "data": {"target_id": 1}}
        ]

        return TestScenario(
            name="Multiple Targets",
            description="Drone encounters multiple targets and manages priorities",
            initial_state=initial_state,
            events=events,
            expected_outcome=MissionOutcome.SUCCESS,
            success_criteria=[
                "All targets detected",
                "Proper priority management",
                "Sequential engagement successful",
                "Mission objectives achieved"
            ]
        )

    def get_all_scenarios(self) -> List[TestScenario]:
        """Return all test scenarios"""
        return [
            self.scenario_1_successful_engagement(),
            self.scenario_2_no_target_found(),
            self.scenario_3_low_fuel_emergency(),
            self.scenario_4_obstacle_encounter(),
            self.scenario_5_return_to_base_command(),
            self.scenario_6_system_failures(),
            self.scenario_7_weather_emergency(),
            self.scenario_8_multiple_targets()
        ]

    def run_scenario_simulation(self, scenario: TestScenario, verbose: bool = True) -> Dict:
        """Simulate running a test scenario"""
        if verbose:
            print(f"\n{'='*60}")
            print(f"RUNNING SCENARIO: {scenario.name}")
            print(f"{'='*60}")
            print(f"Description: {scenario.description}")
            print(f"Expected Outcome: {scenario.expected_outcome.value}")
            print(f"\nSuccess Criteria:")
            for criteria in scenario.success_criteria:
                print(f"  - {criteria}")
            print(f"\n{'='*60}")

        # Simulate the scenario events
        current_time = 0
        results = {
            "scenario_name": scenario.name,
            "events_processed": [],
            "final_state": None,
            "success": False,
            "criteria_met": [],
            "criteria_failed": []
        }

        current_state = scenario.initial_state

        for event in scenario.events:
            if verbose:
                print(f"Time {event['time']:4d}s: {event['type']} - {event.get('data', {})}")

            results["events_processed"].append({
                "time": event["time"],
                "type": event["type"],
                "data": event.get("data", {})
            })

            # Update state based on event (simplified simulation)
            current_time = event["time"]
            current_state.mission_time = current_time

            # Process different event types
            if event["type"] == "fuel_critical":
                current_state.fuel_level = event["data"]["fuel_level"]
            elif event["type"] == "target_detected":
                current_state.target_detected = True
                current_state.target_position = event["data"]["position"]
            elif event["type"] == "obstacle_detected":
                current_state.obstacle_detected = True
                current_state.obstacle_position = event["data"]["position"]
            elif event["type"] == "return_to_base_command":
                current_state.return_to_base_commanded = True
            elif event["type"] in ["sensor_degradation", "communication_failure", "navigation_error"]:
                system = event["data"]["system"]
                status = SystemStatus(event["data"]["status"])
                current_state.system_health[system] = status

        results["final_state"] = current_state

        # Evaluate success criteria (simplified)
        criteria_evaluation = self.evaluate_success_criteria(scenario, results)
        results.update(criteria_evaluation)

        if verbose:
            print(f"\n{'='*60}")
            print(f"SCENARIO RESULTS:")
            print(f"Success: {'' if results['success'] else ''}")
            print(f"Criteria Met: {len(results['criteria_met'])}/{len(scenario.success_criteria)}")
            for criteria in results["criteria_met"]:
                print(f"   {criteria}")
            for criteria in results["criteria_failed"]:
                print(f"   {criteria}")
            print(f"{'='*60}")

        return results

    def evaluate_success_criteria(self, scenario: TestScenario, results: Dict) -> Dict:
        """Evaluate if success criteria were met (simplified evaluation)"""
        criteria_met = []
        criteria_failed = []

        # This is a simplified evaluation - in a real system, you would have
        # more sophisticated logic to evaluate each criteria
        for criteria in scenario.success_criteria:
            # Simplified criteria evaluation based on keywords
            if any(keyword in criteria.lower() for keyword in ["target", "engagement"]):
                if any("engagement_successful" in event["type"] for event in results["events_processed"]):
                    criteria_met.append(criteria)
                else:
                    criteria_failed.append(criteria)
            elif "return" in criteria.lower() or "landing" in criteria.lower():
                if any(event["type"] in ["landed_at_base", "safe_landing", "emergency_landing"]
                       for event in results["events_processed"]):
                    criteria_met.append(criteria)
                else:
                    criteria_failed.append(criteria)
            else:
                # Default to met for demonstration
                criteria_met.append(criteria)

        success = len(criteria_failed) == 0

        return {
            "success": success,
            "criteria_met": criteria_met,
            "criteria_failed": criteria_failed
        }

    def run_all_scenarios(self, verbose: bool = True) -> Dict:
        """Run all test scenarios and return summary results"""
        all_results = []
        scenarios = self.get_all_scenarios()

        print(f"Running {len(scenarios)} test scenarios for drone mission control...")

        for scenario in scenarios:
            result = self.run_scenario_simulation(scenario, verbose)
            all_results.append(result)

        # Generate summary
        total_scenarios = len(all_results)
        successful_scenarios = sum(1 for result in all_results if result["success"])

        summary = {
            "total_scenarios": total_scenarios,
            "successful_scenarios": successful_scenarios,
            "failed_scenarios": total_scenarios - successful_scenarios,
            "success_rate": successful_scenarios / total_scenarios,
            "detailed_results": all_results
        }

        if verbose:
            print(f"\n{'='*80}")
            print(f"OVERALL TEST SUMMARY")
            print(f"{'='*80}")
            print(f"Total Scenarios: {summary['total_scenarios']}")
            print(f"Successful: {summary['successful_scenarios']}")
            print(f"Failed: {summary['failed_scenarios']}")
            print(f"Success Rate: {summary['success_rate']:.1%}")
            print(f"{'='*80}")

        return summary

# Example usage and demonstration
if __name__ == "__main__":
    # Create test scenarios
    test_framework = DroneTestScenarios()

    # Run a single scenario
    scenario = test_framework.scenario_1_successful_engagement()
    single_result = test_framework.run_scenario_simulation(scenario)

    # Run all scenarios
    all_results = test_framework.run_all_scenarios(verbose=False)

    # Print scenario names for reference
    print("\nAvailable Test Scenarios:")
    for i, scenario in enumerate(test_framework.get_all_scenarios(), 1):
        print(f"{i}. {scenario.name}")
        print(f"   {scenario.description}")
```
